```python

import random

#Used to set case values
cases = [5000, 10000, 25000, 50000, 75000,100000, 200000,300000, 400000, 500000, 750000, 1000000]
    
#Used to see which case you've selected
checkList = [] 

#Used to see the value of your case
yourCase = []

playerName = input("Hey, First Identify your self, What is your name? \n")

def main():
    
    #Let's begin by picking off a box from the available counter
    
    myCase = int(input("Choose a Deal Box (1,11): "))

    #Checks to see if the value you've input is within the index range of the the length of cases
    if myCase > 11 or myCase < 1:
        print ("Sorry ", playerName, " the box selected is Out Of Range")
        main()

    else:

        #Used to randomly select a value from the list of Values that was listed above
        Case = random.choice(cases)

        #Removes that Value so it cannot be re-used in future casses
        cases.remove(Case)

        #Adds your case to the list to keep track of your case Value
        yourCase.append(Case)

        #Adds the number of the case you selected to the list
        checkList.append(myCase)

        #Directs you to the new function
        Remove(myCase,checkList)
        
        #Pass the decission to the banker.
        Banker(myCase,checkList)

#This function's purpose is to remove cases as they are selected.
def Remove(myCase,checkList):
    #Determines how many turns you have left until moving on to the Banker() funtion
    turn = int(input(playerName + " How many turns/rounds do you want to go? \n"))

    #A while loop to get all turns Removals done
    while turn != 0:

    #Again just to see which case/box you've selected or not
        caseRemove = int(input("Round " + str(turn)+ ": Choose a Deal box to be removed (1,11): "))
        #Checks to see if the input is within index range
        if caseRemove > 11 or caseRemove < 1:
            print ("Hey, ", playerName, ", Sorry, the box selected is Out Of Range")

        #Checks to see if the number you've input has already been used
        if caseRemove in checkList:
            print ("Sorry ", playerName, " the box, " + str(caseRemove) + " is already taken.")
        else:
            #Generate a random choice 
            Case = random.choice(cases)
            #Remove the case/choice from the list.
            cases.remove(Case)
            #Add the removed case to our checkList
            checkList.append(caseRemove)
            #Reduce the number of turns to go by 1.
            turn -= 1

    #Checks to see if there is one box/case left
    if len(cases) == 1:
        Banker(myCase,checkList)
        Banker(myCase,checkList)

#The Banker() function is to offer you the average amount of what is still inside the boxes/cases.
def Banker(myCase,checkList):
    #Used to find the Average of your deal boxes
    bankOffer = sum(cases)/float(len(cases))
    #Again checks to see if you have 1 case left over
    if len(cases) == 1:

        #Asks you to see if you want your box/case or not
        finalChoice = input("Would you like to keep your case (Y)es or (N)o ? \n")
        #Let's convert our input to upper case only.
        FStnd = finalChoice.upper()
        if FStnd == "Y":
            #This Little peice is to make it when the number is displayed, it won't have brackets around it
            yourCase2 = sum(yourCase)
            print (playerName, " You won UGX: "+ str(yourCase2))
            main()
            
        elif FStnd == "N":
            print (playerName, ", You won UGX: " +str(bankOffer))
            main()

        #If you input something besides n or y then it keeps asking the question until you get it right
        else:
            print ("Sorry, that was an invalid Input, try again")
            Banker(myCase,checkList)

    #this portion is for when the cases are greater than 1
    if len(cases) > 1:

        #We now show the offer the bank is making to the player.
        decide = input("The Banker Offered UGX: " + str(bankOffer) + " Deal, or No Deal? (D)eal or (N)o Deal \n" )
        offerDecision = decide.upper()

        if offerDecision == "D":
            print ("Congs, ", playerName + ", You have won: UGX " + str(bankOffer))
            #Show him what he would have won from the cases selected.
            yourCase2 = sum(yourCase)

            print("Your case contained: UGX " + str(yourCase2) )
        elif offerDecision == "N":
            #If the player doesn't take the amount offered by the bank.
            #We allow the user to choose another box if he has more turns to play
            yourCase2 = sum(yourCase)
            Remove(myCase,checkList)
            print("You won UGX: " + str(yourCase2) + ", the bank was offering UGX: " + str(bankOffer) )
            
        else:
            print("Sorry, that was an invalid Input") 
            #Refresh the game.
            Banker(myCase,checkList)
main()
```
